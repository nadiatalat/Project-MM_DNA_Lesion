```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def repair_probability(z, steepness, midpoint):
    """
    Logistic function to calculate repair probability.

    Parameters:
    - z         : time intervals
    - steepness : controls how steep the curve is
    - midpoint  : the point where the probability is 0.5

    Returns:
    - An array of probabilities ranging from 0 to 1.
    """
    return 1 / (1 + np.exp(-steepness * (z - midpoint)))
```

```python
# Time interval values (x-axis)
z_values = np.linspace(0, 10, 500)  # From 0 to 10, with 500 points

# Parameter values to test
steepness_values = [0.5, 1.0, 2.0]   # Controls how fast the curve rises
midpoint_values = [2.0, 4.0, 6.0]    # The z-value where P_r(z) â‰ˆ 0.5
```

```python
# Rows = different steepness values, Columns = different midpoints
fig, axes = plt.subplots(
    len(steepness_values),
    len(midpoint_values),
    figsize=(15, 10),
    sharex=True,
    sharey=True
)

for i, steepness in enumerate(steepness_values):
    for j, midpoint in enumerate(midpoint_values):
        ax = axes[i, j]  # Access subplot at row i, column j
        P_r_values = repair_probability(z_values, steepness, midpoint)  # Calculate probabilities

        # Plot the logistic curve
        ax.plot(z_values, P_r_values, label=f'steepness={steepness}, midpoint={midpoint}')
        ax.set_title(f'steepness={steepness}, midpoint={midpoint}')
        ax.grid(True)
        ax.set_ylim([0, 1])  # Probability range

        # Label x-axis only on the bottom row
        if i == len(steepness_values) - 1:
            ax.set_xlabel('Time interval z')

        # Label y-axis only on the first column
        if j == 0:
            ax.set_ylabel('Repair Probability')
```
